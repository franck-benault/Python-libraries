# Plot functions

## Simple examples

```python
from sympy import *

x=symbols('x')
f= 2*x+7
plot(f)
```

```python
from sympy import *

x=symbols('x')
f2= 5*x**2+7*x-4
plot(f2)
```

## 3D diagram

```python
from sympy import *
from sympy.plotting import plot3d

x, y=symbols('x y')
f= 5*x+3*y
plot3d(f)
```

# Limits

```python
from sympy import *

x=symbols('x')
f=1/x
print(limit(f,x,oo))
```

```python
x=symbols('x')
f=(1+1/x)**x
print(limit(f,x,oo))
print(limit(f,x,oo).evalf())
```
